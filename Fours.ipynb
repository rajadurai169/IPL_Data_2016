```python
# Importing Pandas as Alice Name 'pd'
import pandas as pd
score = pd.read_csv(r"Batsman_Scored.csv")
score.info()
```

```python
ball_by_ball = pd.read_csv(r"Ball_by_Ball.csv")
ball_by_ball = ball_by_ball.rename(columns={"Striker":"Player_Id"})
ball_by_ball = ball_by_ball[["Match_Id","Innings_No","Player_Id"]]
ball_by_ball.tail()
```

```python
ball_score = pd.merge(score,ball_by_ball,on=["Match_Id","Innings_No"],how="inner")
ball_score.head()
ball_score.drop_duplicates()
```

```python
player = pd.read_csv(r"Player.csv")
player = player[["Player_Id","Player_Name"]]
ball_score_player = pd.merge(ball_score,player,on ="Player_Id",how="inner")
ball_score_player.head()
```

```python
match = pd.read_csv(r"Match.csv")
match= match[["Match_Id","Match_Date"]]
match.head()
```

```python
match_ball_score = pd.merge(ball_score_player,match,on="Match_Id",how="inner")
match_ball_score.tail()
```

```python
match_ball_score.drop_duplicates()
```

```python
match_ball_score["Year"] = match_ball_score["Match_Date"].str.slice(0,4)
match_ball_score.head()
```

```python
match_ball_score1 = match_ball_score.groupby(["Player_Id","Player_Name","Match_Id","Year"])["Runs_Scored"].count()
match_ball_score1.head(5)
```
