```python
# Importing pandas with alice name pd
import pandas as pd

# Reading "Ball_by_Ball.csv" File for Our Analysis in variable "ball_by_ball"
ball_by_ball =pd.read_csv(r"D:\pandas_learing\IPL_Project\Ball_by_Ball.csv")

# Changing "Striker" into "Player_Id" Because Both are Same 
ball_by_ball = ball_by_ball.rename(columns={"Striker":"Player_Id"})

# Printing ball_by_ball for to see the Outcome Data
ball_by_ball.head()
```

```python
# Reading "Batsman_Scored.csv" File for Our Analysis in variable "score"
score =pd.read_csv(r"D:\pandas_learing\IPL_Project\Batsman_Scored.csv")

# Printing score for to see the Outcome Data
score.head()
```

```python
# Mergeing "ball_by_ball" with "score" with comes in variable ball_score
ball_score =pd.merge(ball_by_ball,score,on=['Match_Id','Over_Id','Ball_Id','Innings_No'],how='inner')

# Printing ball_score for to see the Outcome Data
ball_score.head()
```

```python
# Reading "Player.csv" File for Our Analysis in variable "player"
player = pd.read_csv(r"D:\pandas_learing\IPL_Project\Player.csv")

# Printing player for to see the Outcome Data
player.head()
```

```python
# Mergeing ball_score with player with come "Player_Id"
ball_score_player_name=pd.merge(ball_score,player,on="Player_Id",how="inner")

# Printing ball_Score_player_name for to see the Outcome Data
ball_score_player_name.head()
```

```python
# Reading "Match.csv" File for Our Analysis in variable match
match =pd.read_csv(r"D:\pandas_learing\IPL_Project\Match.csv")

# Printing match for to see the Outcome Data
match.head()
```

```python
# Storing Wanten Column in valiable Match_want
match_want = match[['Match_Id','Match_Date']]

# Printing match_want for to see the Outcome Data
match_want.head()
```

```python
# Mergeing ball_score_player_name with match_want
final = pd.merge(ball_score_player_name,match,on="Match_Id",how='left')

# Printing final for to see the Outcome Data
final.head()
```
