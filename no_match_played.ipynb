```python

# Importing Required Modules Pandas as alice Name = 'pd'
import pandas as pd 

# Reading Match.csv File
match_details = pd.read_csv(r"Match.csv")
#match_details.head(3)
match_details = match_details[["Match_Id","Match_Date"]]
match_details.head()
```

```python
player = pd.read_csv(r"Player.csv")
player.head()
```

```python
match = pd.read_csv(r"Player_Match.csv")
match.head()
```

```python
player_match = pd.merge(player,match,on="Player_Id",how ="inner")
player_match.head(10)
```

```python
match_details.head()
```

```python
player_details = pd.merge(player_match,match_details,on="Match_Id",how="inner")
player_details.head(10)
```

```python
player_details.drop_duplicates()
player_details.head()
```

```python
player_details = player_details[["Player_Id","Player_Name","Match_Id","Match_Date"]]
player_details.head(10)
```

```python
player_details = player_details.reset_index()
player_details.head(10)
```

```python
player_details["Match_Year"] = player_details["Match_Date"].str.slice(0,4)
player_details.head(10)
player_details_1= player_details[player_details["Match_Year"]=="2016"]
player_details_1.head(10)
```

```python
match_played= player_details_1.groupby(["Player_Id",'Player_Name'])['Match_Id'].nunique().reset_index()
match_played.head()
```

```python
match_played = match_played.rename(columns={"Match_Id":"Match_Played"})
match_played =match_played.sort_values("Match_Played",ascending=False)
match_played.head()
```

```python
final_2016 = pd.read_csv(r"Final_2016.csv")
final_2016.head()
```

```python
no_match_2016 = pd.merge(match_played,final_2016,on="Player_Name",how="left")
no_match_2016.head()
```

```python
no_match_2016 = no_match_2016.sort_values("Runs_Scored",ascending=False)
no_match_2016.head()
```

```python
no_match_2016 = no_match_2016[["Player_Id","Player_Name","Match_Played","Runs_Scored"]]
no_match_2016.head(10)
```

```python
no_match_2016.to_csv("no_match_2016.cvs")
```
