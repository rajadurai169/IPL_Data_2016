```python
import pandas as pd
```

```python
ball_by_ball = pd.read_csv(r"D:\pandas_learing\IPL_Project\Ball_by_Ball.csv")
ball_score=ball_by_ball.rename(columns={"Striker":"Player_Id"})
ball_score.head()
```

```python
score = pd.read_csv(r"D:\pandas_learing\IPL_Project\Batsman_Scored.csv")
score.head()
```

```python
ball_score = pd.merge(ball_score,score,on =["Match_Id","Over_Id","Ball_Id","Innings_No"],how="inner")
ball_score.head()
```

```python
player =pd.read_csv(r"D:\pandas_learing\IPL_Project\Player.csv")
player.head()
```

```python
match = pd.read_csv(r"D:\pandas_learing\IPL_Project\Match.csv")
match.head()
```

```python
player_match = pd.merge(ball_score,player,on="Player_Id",how="inner")
player_match.head()
```

```python
player_match.groupby(['Player_Name'])['Runs_Scored'].sum()
```

```python
match_new = match[["Match_Id","Match_Date"]]
match_new = match_new.drop_duplicates()
match_new.head()
```

```python
final = pd.merge(player_match,match_new,on="Match_Id",how="left")
```

```python
final.head()
```

```python
final["Year"]=final["Match_Date"].str.slice(0,4)
final.head()
```

```python
final1=final.groupby(["Player_Id",'Player_Name','Year'],as_index=False)['Runs_Scored'].sum()
final1.head()
```

```python
final1_2016=final1[final1['Year']=='2016']
final1_2016=final1_2016.sort_values("Runs_Scored",ascending=False)
```
