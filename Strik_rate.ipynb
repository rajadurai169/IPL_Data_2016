```python
# Importing pandas as alice name pd

import pandas as pd

# Reading Final2_2016.csv File to Analysis strik rate 

final2_2016 = pd.read_csv(r"Final2_2016.csv")
final2_2016 = final2_2016[["Player_Id","Player_Name","Year","Runs_Scored"]]
final2_2016.head()
```

```python
# reading Ball_by_Ball.csv File to Combine Final_2016.csv for Analysis

ball_by_ball = pd.read_csv(r"Ball_by_Ball.csv")
# Changing Stricker Column into Player_Id Because Both are Same
ball_by_ball.rename(columns={"Striker":"Player_Id"},inplace = True)
ball_by_ball=ball_by_ball.drop(columns = ["Over_Id","Innings_No","Team_Batting","Team_Bowling","Striker_Batting_Position","Non_Striker","Bowler"],axis=1)
ball_by_ball.head(2)
```

```python

merge_file = pd.merge(final2_2016,ball_by_ball,on="Player_Id",how='inner')
merge_file.head(2)
```

```python
no_ball = merge_file.groupby(["Player_Name"])["Ball_Id"].count()
no_ball.tail()
```
